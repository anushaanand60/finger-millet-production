# **Part A: Cobb-Douglas Production Function**

The Cobb-Douglas production function is given by:

$Y = A X_1^{\alpha_1} X_2^{\alpha_2} \dots X_K^{\alpha_K}$

Where:
-  Y : Output (production of Finger Millet, `production1000tonnes`).
-  A : Total factor productivity (technology level, to be estimated as the intercept in the log-linear model).
- $ X_k $: Inputs (e.g., area, irrigation, fertilizers).
- $ \alpha_k $: Elasticity of output with respect to the $  k  $ -th input (to be estimated as coefficients).

To estimate this, we take the natural log of both sides to convert it into a log-linear form:

$ln(Y) = \ln(A) + \alpha_1 \ln(X_1) + \alpha_2 \ln(X_2) + \dots + \alpha_K \ln(X_K)$


We’ll include environmental controls (rainfall and soil salinity) as additional inputs, but without taking their logs, as specified.

#### **Define the Inputs and Output**
- **Output ($ Y  $)**: `production1000tonnes` (log-transformed as `ln_production`).
- **Inputs ($ X_k  $) to be log-transformed**:
  - `area1000hectares`: Total cropped area (proxy for land input).
  - `irrigatedarea_new`: Irrigated area, redefined as `irrigatedarea1000hectares + 1` to handle zeros.
  - `unirrigatedarea`: Unirrigated area, defined as `area1000hectares - irrigatedarea_new`.
  - `nitrogenconsumptiontonnes`: Nitrogen fertilizer consumption.
  - `phosphateconsumptiontonnes`: Phosphate fertilizer consumption.
  - `potashconsumptiontonnes`: Potash fertilizer consumption.
- **Environmental Controls (not log-transformed)**:
  - `kharif_rainfall_mm`: Rainfall during the Kharif season.
  - `salinity_alkalinity_percentage`: Soil salinity/alkalinity percentage.

The log-linear model will be:

$$
\begin{aligned}
\ln(\text{production1000tonnes}) &= \beta_0 + \beta_1 \ln(\text{irrigatedarea\_new}) + \beta_2 \ln(\text{unirrigatedarea}) + \beta_3 \ln(\text{nitrogenconsumptiontonnes}) + \beta_4 \ln(\text{phosphateconsumptiontonnes}) \\
&\quad + \beta_5 \ln(\text{potashconsumptiontonnes}) + \beta_6\, \text{kharif\_rainfall\_mm} + \beta_7\, \text{salinity\_alkalinity\_percentage} + \epsilon
\end{aligned}
$$

Where:
- $ \beta_0 = \ln(A) $: Intercept (log of total factor productivity).
- $ \beta_1, \beta_2, \dots, \beta_5 $: Elasticities of production with respect to each log-transformed input.
- $ \beta_6, \beta_7 $: Coefficients for the environmental controls (not log-transformed).
- $ \epsilon $: Error term.
Let’s update the data preparation steps for the quadratic production function model based on the data handling approach used for the Cobb-Douglas model. ### **1. Are There Any Outliers and/or Influential Observations?**
#### **Methodology**
- **Outliers**: These are observations with extreme values in the dependent variable (`production1000tonnes`) or independent variables that deviate significantly from the regression line. We can identify them using:
  - **Studentized Residuals**: Adjusted residuals that account for leverage, with values > 3 suggesting outliers.

- **Influential Observations**: These are points that disproportionately affect the regression coefficients or fitted values. We will use:
  - **DFBETA**: Measures the change in each coefficient when the i-th observation is removed. Values > 2/sqrt(n) (where n = 72, so 2/sqrt(72) ≈ 0.236) or > 1 for individual coefficients indicate influence.
  - **DFFITS**: Measures the change in the fitted value for the i-th observation when it is removed. Values > 2*sqrt(k/n) (where k is the number of predictors, e.g., 7, so 2*sqrt(8/72) ≈ 0.667) indicate influence.




```python
# Get residuals and studentized residuals
influences = model_cd.get_influence()
student_resid = influences.resid_studentized

# Identify outliers (threshold: |studentized residual| > 3)
outliers = np.where(np.abs(student_resid) > 3)[0]
print(f"Outlier indices (0-based): {outliers}")
print(f"Number of outliers: {len(outliers)}")
print(f"Studentized residuals of outliers: {student_resid[outliers]}")

# Optional: Plot residuals
import matplotlib.pyplot as plt
plt.scatter(range(len(student_resid)), student_resid)
plt.axhline(y=3, color='r', linestyle='--')
plt.axhline(y=-3, color='r', linestyle='--')
plt.xlabel('Observation Index')
plt.ylabel('Studentized Residuals')
plt.title('Studentized Residuals Plot')
plt.show()
```

### **Why the Threshold of 3?**

The choice of a threshold of 3 for studentized residuals is based on statistical theory and practical conventions:

- **Theoretical Basis**:
  - Under the assumption of normally distributed errors, studentized residuals follow a t-distribution with $n - k - 1$ degrees of freedom (where $ = 72$is the number of observations, and $k = 7$ is the number of predictors in the model, so df = 72 - 7 - 1 = 64).
  - For a t-distribution with 64 degrees of freedom, the critical value at a 99% confidence level (two-tailed) is approximately 2.66. A threshold of 3 is slightly more conservative and aligns with the common rule of thumb for detecting outliers in regression, where |residual| > 3 standard errors is considered extreme.

- **Practical Convention**:
  - In regression diagnostics, a studentized residual exceeding 3 is often used as a cutoff to flag potential outliers because it corresponds to a p-value of roughly 0.001 (for a two-tailed test), indicating a very low probability of occurring under the null hypothesis of normality.
  - This threshold balances sensitivity (detecting true outliers) and specificity (avoiding false positives), especially with moderate sample sizes like 72.

- **Adjustment for Sample Size**:
  - With a larger sample (e.g., > 100), a threshold of 2.5 or 2 might be considered due to increased power to detect smaller deviations. For the sample size of 72, 3 is appropriate.

- **Contextual Consideration**:
  - Since the model has a high R-squared (0.860) and the residuals don’t exceed 3, the threshold of 3 is justified.

## **2. What Are the Collinearity Diagnostics?**

### Methodology
- **Variance Inflation Factor (VIF)**: Measures the increase in variance due to multicollinearity:
  $
  \text{VIF} = \frac{1}{1 - R_k^2}
  $
  Where $R_k^2$ is the $R^2$ from regressing $x_k$ on other predictors. VIF > 10 indicates high multicollinearity.
- **Condition Index (CI)**: Defined as:
  $
  CI = \sqrt{\frac{\lambda_{\max}}{\lambda_{\min}}}
  $
  Where $lambda_{\max}$ and $\lambda_{\min}$ are the largest and smallest eigenvalues of $X'X$. CI > 20 (or 30) suggests multicollinearity.

# **Part B: Quadratic Production Function**

The quadratic production function models output ($ Y $) as a function of inputs ($ X_k; k=1,2,\ldots,K $) in their original levels, incorporating squared terms to account for diminishing returns beyond a certain threshold. The general form is:

$$
Y = \beta_0 + \beta_1 X_1 + \beta_2 X_1^2 + \beta_3 X_2 + \beta_4 X_2^2 + \cdots + \beta_K X_K + \beta_{K+1} X_K^2 + \epsilon
$$

Where:
- $ Y $: Output (e.g., `production1000tonnes`).
- $ \beta_0 $: Intercept (baseline output level).
- $ X_k $: Input variables in their original units (e.g., area, irrigation, fertilizers).
- $ \beta_1, \beta_3, \ldots, \beta_K $: Linear coefficients for each input.
- $ \beta_2, \beta_4, \ldots, \beta_{K+1} $: Coefficients for the squared terms, capturing diminishing effectiveness of inputs.
- $ \epsilon $: Error term.

The inclusion of squared terms allows the model to reflect non-linear relationships, where the marginal effect of an input decreases as its level increases. Given the assumption that input use is always positive, we won’t need to adjust for zero or negative values, simplifying the transformation process.

#### **Define the Inputs and Output**
- **Output ($ Y $)**: `production1000tonnes` (used in its original level form, not log-transformed).
- **Inputs ($ X_k $) with Linear and Squared Terms**:
  - `irrigatedarea_capped`: Capped irrigated area (in 1000 hectares), with a squared term `irrigatedarea_capped^2`.
  - `unirrigatedarea`: Unirrigated area (in 1000 hectares), with a squared term `unirrigatedarea^2`.
  - `nitrogenconsumptiontonnes`: Nitrogen fertilizer consumption (in tonnes), with a squared term `nitrogenconsumptiontonnes^2`.
  - `phosphateconsumptiontonnes`: Phosphate fertilizer consumption (in tonnes), with a squared term `phosphateconsumptiontonnes^2`.
  - `potashconsumptiontonnes`: Potash fertilizer consumption (in tonnes), with a squared term `potashconsumptiontonnes^2`.
- **Environmental Controls (linear terms only, no squared terms)**:
  - `kharif_rainfall_mm`: Rainfall during the Kharif season.
  - `salinity_alkalinity_percentage`: Soil salinity/alkalinity percentage.

The proposed quadratic model will be:

$$
\begin{aligned}
\text{production1000tonnes} = \beta_0 + \beta_1 (\text{irrigatedarea\_capped}) + \beta_2 (\text{irrigatedarea\_capped}^2) + \beta_3 (\text{unirrigatedarea}) + \beta_4 (\text{unirrigatedarea}^2) \\
+ \beta_5 (\text{nitrogenconsumptiontonnes}) + \beta_6 (\text{nitrogenconsumptiontonnes}^2) 
+ \beta_7 (\text{phosphateconsumptiontonnes}) \\ 
+ \beta_8 (\text{phosphateconsumptiontonnes}^2) + \beta_9 (\text{potashconsumptiontonnes}) + \beta_{10} (\text{potashconsumptiontonnes}^2) \\
+ \beta_{11} (\text{kharif\_rainfall\_mm}) + \beta_{12} (\text{salinity\_alkalinity\_percentage}) + \epsilon
\end{aligned}
$$

Where:
- $ \beta_0 $: Intercept (baseline production level).
- $ \beta_1, \beta_3, \beta_5, \beta_7, \beta_9 $: Linear effects of each input.
- $ \beta_2, \beta_4, \beta_6, \beta_8, \beta_{10} $: Effects of the squared terms, indicating diminishing returns if negative.
- $ \beta_{11}, \beta_{12} $: Coefficients for the environmental controls.
- $ \epsilon $: Error term.

## **Question 7:**
To test the hypothesis of input complementarity between irrigation and fertilizer application for Finger Millet—specifically that additional fertilizer application is more effective with irrigation than without—we need to assess whether the marginal productivity of fertilizers increases in the presence of irrigation. In economic terms, input complementarity implies that the marginal product of one input (e.g., fertilizer) increases with the level of another input (e.g., irrigation). This can be modeled by including interaction terms between irrigation and fertilizer inputs in the quadratic production function, allowing us to test whether the interaction effect is positive and significant.

### **Theoretical Framework**
In a quadratic production function, the marginal product of a fertilizer input (e.g., `nitrogenconsumptiontonnes`) is:
$\frac{\partial Y}{\partial (\text{nitrogenconsumptiontonnes})} = \beta_5 + 2 \beta_6 \cdot \text{nitrogenconsumptiontonnes}$

Input complementarity with irrigation (e.g., `irrigatedarea_capped`) would suggest that this marginal product increases as `irrigatedarea_capped` increases. This can be captured by adding interaction terms such as:
$\text{irrigatedarea\_capped} \times \text{nitrogenconsumptiontonnes}, \quad \text{irrigatedarea\_capped} \times \text{nitrogenconsumptiontonnes}^2, $
and similar terms for other fertilizers (phosphate and potash). A positive and significant coefficient on the linear interaction term (e.g., $ \text{irrigatedarea\_capped} \times \text{nitrogenconsumptiontonnes} $) would indicate that the marginal effect of fertilizer increases with irrigation, supporting complementarity.

### **Hypotheses**
For each fertilizer input, we define the hypotheses based on the interaction term with irrigation. Let’s focus on the linear interaction term $ (\text{irrigatedarea\_capped} \times \text{nitrogenconsumptiontonnes}) $ as the primary test (we can extend to squared terms if needed):
- **Null Hypothesis ($ H_0 $)**: The coefficient of the interaction term $\beta_{\text{interaction}} = 0$ (No complementarity; the marginal effect of fertilizer is independent of irrigation level).
- **Alternate Hypothesis ($ H_a $)**: $\beta_{\text{interaction}} > 0$ (Complementarity; the marginal effect of fertilizer increases with irrigation level).

We will conduct a one-sided t-test for each fertilizer-irrigation interaction, using the p-value to assess significance in the upper tail.

### **Augmenting the Model**
To test this hypothesis, we augment the existing quadratic model by adding interaction terms between `irrigatedarea_capped` and each fertilizer variable (`nitrogenconsumptiontonnes`, `phosphateconsumptiontonnes`, `potashconsumptiontonnes`). The augmented model becomes:

$$
\begin{aligned}
\text{production1000tonnes} = & \beta_0 + \beta_1 (\text{irrigatedarea\_capped}) + \beta_2 (\text{irrigatedarea\_capped}^2) + \beta_3 (\text{unirrigatedarea}) + \beta_4 (\text{unirrigatedarea}^2) \\
& + \beta_5 (\text{nitrogenconsumptiontonnes}) + \beta_6 (\text{nitrogenconsumptiontonnes}^2) + \beta_7 (\text{phosphateconsumptiontonnes}) \\

& + \beta_{11} (\text{irrigatedarea\_capped} \times \text{nitrogenconsumptiontonnes}) + \beta_{12} (\text{irrigatedarea\_capped} \times \text{phosphateconsumptiontonnes}) \\
& + \beta_{13} (\text{irrigatedarea\_capped} \times \text{potashconsumptiontonnes}) + \beta_{14} (\text{kharif\_rainfall\_mm}) + \beta_{15} (\text{salinity\_alkalinity\_percentage}) + \epsilon
\end{aligned}
$$

- The interaction terms $ (\beta_{11}, \beta_{12}, \beta_{13}) $ are the key variables to test for complementarity.
- Optionally, we could include interactions with squared fertilizer terms (e.g., $\text{irrigatedarea\_capped} \times \text{nitrogenconsumptiontonnes}^2 $) , but the linear interaction is a standard first step to assess complementarity.
